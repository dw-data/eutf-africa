```python
ref = pd.read_csv('../data/country-and-continent-codes-list.csv', delimiter=';')
```

```python
ref.head()
```

```python
df_merged = df.merge(ref[['ISO-alpha3 code', 'continent', 'M49 code']], left_on = 'Country of origin (ISO)', right_on='ISO-alpha3 code', how = 'left')
```

```python
len(df_merged)
```

```python
len(df)
```

```python
df_merged.head()
```

```python
df_merged[df_merged['continent'].isnull()]
```

```python
df_africa = df_merged[df_merged['continent']=='Africa']
```

```python
df_africa.head()
```

```python
df_africa.columns
```

```python
df_africa = df_africa[["Year", "Country of origin", "Country of origin (ISO)",
       "Refugees under UNHCR's mandate", "Asylum-seekers",
       "IDPs of concern to UNHCR",
       "Stateless persons", "Others of concern", "M49 code"]]
```

```python
df_africa.columns =(['Year', 'Origin_Country', 'ISO3',
       'Refugees', 'Asylum-seekers',
       'IDPs', 'Stateless', 'Others',
       'ISO_numeric'])
```

```python
df_africa.head()
```

```python
df_africa_melted = df_africa.melt(id_vars=['Origin_Country','ISO3','ISO_numeric','Year'])
```

```python
df_africa_melted.head()
```

```python
alt.Chart(df_africa_melted).mark_line().encode(
    x='Year',
    y='value',
    color='variable',
    tooltip=['Year', 'variable', 'value']
).properties(
    width=100,
    height=100
).facet(
    facet='Origin_Country:N',
    columns=4
)
```

```python
df_africa.head()
```

```python
df_africa['all_of_concern'] = df_africa['Refugees']+df_africa['Asylum-seekers']+df_africa['IDPs']+df_africa['Stateless']+df_africa['Others']
```

```python
df_africa.head()
```

```python
df_africa.columns
```

```python
df_africa_simplified = df_africa[['Year', 'Origin_Country', 'ISO3', 'ISO_numeric', 'all_of_concern']]
```

```python
df_africa_simplified.head()
```

```python
df_africa_simplified['Year']=pd.to_datetime(df_africa_simplified['Year'], format='%Y')
```

```python
AoC_Africa_yearly=df_africa_simplified.groupby(by='Year')['all_of_concern'].sum().reset_index()
```

```python
alt.Chart(AoC_Africa_yearly, title='UNHCR people of concern with African origin').mark_bar(width=20).encode(
    x='Year',
    y='all_of_concern',
)
```

```python
edf=pd.read_csv('../data/EUTF-country-budget.csv')
edf.head()
```

```python
df_africa_simplified_m = df_africa_simplified.merge(edf, left_on='ISO3', right_on='ISO', how='outer')
```

```python
df_africa_simplified_m.head()
```

```python
AoC_eutf_countries = df_africa_simplified_m[df_africa_simplified_m['country'].notnull()]
AoC_eutf_countries.head()
```

```python
len(AoC_eutf_countries.Origin_Country.value_counts())
```

```python
AoC_eutf_yearly=AoC_eutf_countries.groupby(by='Year')['all_of_concern'].sum().reset_index()
AoC_eutf_yearly
```

```python
africa = alt.Chart(AoC_Africa_yearly, title='UNHCR people of concern with African origin (blue) and from EUTF recipient countries (orange)').mark_bar(width=20).encode(
    x='Year',
    y='all_of_concern',
)

eutf_countries = alt.Chart(AoC_eutf_yearly).mark_bar(width=7, color = 'orange').encode(
    x='Year',
    y='all_of_concern',
)

combined = africa + eutf_countries
combined
```
